```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
import sklearn
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#Data Collection
data = pd.read_csv('/content/framingham.csv')
data
```

```python
print(data.head())
print(data.describe())
```

```python
print(data.info())
print(data.isna().sum()) #if null
```

```python
data.shape
```

```python
data = data.drop("education",axis='columns')
data.head()
```

```python
missing = [var for var in data.columns if data[var].isnull().sum()>0]
missing
```

```python
#how many missing values and in percentage
total_missing_values =0
for col in missing:
  total_missing_values+=data[col].isnull().sum()
total_missing_values
```

```python
#We have 540 missing rows so let's check it's percentage against data
percent_missing_rows = (total_missing_values/data.shape[0])*100
print(percent_missing_rows)
```

```python
data = data.dropna()
data.shape
```

```python
data.head()
```

```python
print(data.isna().sum())
```

```python
#X is independent variable
#y is dependent variable
Y = data['TenYearCHD']
X = data.drop('TenYearCHD',axis=1)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
X[10,:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
Logistic_regression = LogisticRegression()
Logistic_regression.fit(X_train,y_train)
test_prediction = Logistic_regression.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
test_conf_matrix = confusion_matrix(y_test, test_prediction)
test_conf_matrix
```

```python
test_acc_score = accuracy_score(y_test, test_prediction)
test_acc_score
```

```python
test_class_report = classification_report(y_test,test_prediction)
print(test_class_report)
```

Check for overfitting & underfitting


```python
train_prediction = Logistic_regression.predict(X_train)
train_acc_score = accuracy_score(y_train, train_prediction)
train_acc_score
```

```python
print('Training set score: {:.4f}'.format(train_acc_score))

print('Test set score: {:.4f}'.format(test_acc_score))
```
